```python
import pandas as pd
```

```python
movies = pd.read_csv("movies.csv")
movies
movies['year'] = ''
```

```python
# movies['year'] = movies['title'][-5:-1]
for i in range(0, len(movies)):
    movies['year'][i] = movies['title'][i][-5:-1]
    movies['title'][i] = movies['title'][i][:-7]
    movies['title'][i] = movies['title'][i].replace(", The", " [The]").replace(",", "")
movies
```

```python
movies.head(50)
```

```python
links = pd.read_csv("links.csv")
ratings = pd.read_csv("ratings.csv")
tags = pd.read_csv("tags.csv")
```

```python
ratings.head()
```

```python
movies = movies.join(links.set_index('movieId'), on='movieId')
movies = movies.join(ratings.set_index('movieId'), on='movieId')
movies = movies.join(tags.set_index('movieId'), on=('movieId', 'userId'))
```
